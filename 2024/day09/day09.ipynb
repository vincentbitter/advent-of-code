```python
from libpy import readfile

#filename = "2024/day09/input/sample.txt"
filename = "2024/day09/input/input.txt"
diskmap = readfile.lines(filename)[0]

file_blocks = []
free_blocks = []

index = 0
for i,c in enumerate(diskmap):
    if i % 2 == 0:
        file_blocks.append((index, int(c), int(i/2)))
    elif c != '0':
        free_blocks.append((index, int(c)))
    index += int(c)
        
print(file_blocks)
print(free_blocks)
```

```python
reversed_file_blocks = sorted(file_blocks[1:], key=lambda x: x[0], reverse=True)
available_free_blocks = free_blocks.copy()
defragmented_file_blocks = []
defragmented_file_blocks.append(file_blocks[0])
while len(reversed_file_blocks) > 0 and len(available_free_blocks) > 0:
    file_block = reversed_file_blocks[0]
    free_block = available_free_blocks[0]
    left = free_block[1] - file_block[1]
    available_free_blocks.remove(free_block)
    if left == 0:
        defragmented_file_blocks.append((free_block[0], file_block[1], file_block[2]))
        reversed_file_blocks = reversed_file_blocks[1:]
    elif left > 0:
        defragmented_file_blocks.append((free_block[0], file_block[1], file_block[2]))
        available_free_blocks.insert(0, (free_block[0] + file_block[1], left))
        reversed_file_blocks = reversed_file_blocks[1:]
    else:
        defragmented_file_blocks.append((free_block[0], free_block[1], file_block[2]))
        reversed_file_blocks[0] = (file_block[0], left * -1, file_block[2])
    
    next_index = defragmented_file_blocks[-1][0] + defragmented_file_blocks[-1][1]
    while len(reversed_file_blocks) > 0 and reversed_file_blocks[-1][0] == next_index:
        defragmented_file_blocks.append(reversed_file_blocks.pop())
        next_index = defragmented_file_blocks[-1][0] + defragmented_file_blocks[-1][1]

score = 0
for block in defragmented_file_blocks:
    for i in range(block[1]):
        score += (block[0] + i) * block[2]
print(score)
```

```python
reversed_file_blocks = sorted(file_blocks, key=lambda x: x[0], reverse=True)
available_free_blocks = free_blocks.copy()
defragmented_file_blocks = []

for block in reversed_file_blocks:
    for free_block_index in range(len(available_free_blocks)):
        free_block = available_free_blocks[free_block_index]
        if free_block[0] > block[0]:
            break
        
        left = free_block[1] - block[1]
        if left >= 0:
            defragmented_file_blocks.append((free_block[0], block[1], block[2]))
            if left == 0:
                available_free_blocks.remove(free_block)
            else:
                available_free_blocks[free_block_index] = (free_block[0] + block[1], left)
            break
        
reversed_file_blocks = [b for b in reversed_file_blocks if b[2] not in [b2[2] for b2 in defragmented_file_blocks]]

score = 0
for block in defragmented_file_blocks + reversed_file_blocks:
    for i in range(block[1]):
        score += (block[0] + i) * block[2]
print(score)
```
