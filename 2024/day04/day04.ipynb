```python
from libpy import readfile

#filename = "2024/day04/input/sample.txt"
filename = "2024/day04/input/input.txt"
lines = readfile.lines(filename)

print(lines)
```

Part A

```python
height = len(lines)
width = len(lines[0])

def is_mas(x, y, d_x, d_y):
    if x + d_x * 3 < 0 or x + d_x * 3 >= width or y + d_y * 3 < 0 or y + d_y * 3 >= height:
        return False
    return lines[y+d_y][x+d_x] + lines[y+d_y*2][x+d_x*2] + lines[y+d_y*3][x+d_x*3] == "MAS"

score = 0
for x in range(width):
    for y in range(height):
        if lines[y][x] == 'X':
            for d_x in range(-1, 2):
                for d_y in range(-1, 2):
                    if is_mas(x, y, d_x, d_y):
                        score += 1

print(score)
```

Part B

```python
height = len(lines)
width = len(lines[0])

def is_a(a_x, a_y):
    return lines[a_y][a_x] == 'A'

def is_m(m_x, m_y):
    return lines[m_y][m_x] == 'M'

def is_s(s_x, s_y):
    return lines[s_y][s_x] == 'S'

def is_m_and_s(x1, y1, x2, y2):
    return (is_m(x1, y1) and is_s(x2, y2)) or (is_m(x2, y2) and is_s(x1, y1))

score = 0
for x in range(width):
    for y in range(height):
        if x > 0 and x < width - 1 and y > 0 and y < height - 1 and is_a(x,y):
            if is_m_and_s(x-1,y-1,x+1,y+1) and is_m_and_s(x-1,y+1,x+1,y-1):
                score += 1

print(score)
```
