```python
from libpy import readfile

#filename = "2024/day05/input/sample.txt"
filename = "2024/day05/input/input.txt"
lines = readfile.lines(filename)

readrules = True
rules = {}
updates = []
for line in lines:
    if line == '':
        readrules = False    
    elif readrules:
        parts = line.split('|')
        p1 = int(parts[0])
        if p1 not in rules.keys():
            rules[p1] = []
        rules[p1].append(int(parts[1]))
    else:
        updates.append([int(x) for x in line.split(',')])
        
print(rules)
print(updates)
```

```python
def any_item_in(source, target):
    for item in source:
        if item in target:
            return True
    return False

def valid_update(update):
    processed = []
    for value in update:
        if value in rules.keys() and any_item_in(rules[value], processed):
            return False
        processed.append(value)
    return True

def first(needles, haystack):
    lowestIndex = None
    for needle in needles:
        if needle in haystack:
            index = haystack.index(needle)
            if lowestIndex == None or index < lowestIndex:
                lowestIndex = index
    return lowestIndex
```

Part A

```python
score = 0
for update in updates:
    if valid_update(update):
        score += update[len(update)//2]
        
print(score)
```

Part B

```python
score = 0
for update in updates:
    if not valid_update(update):
        update2 = []
        for value in update:
            if value in rules.keys():
                index = first(rules[value], update2)
                if index != None:
                    update2.insert(index, value)
                    continue
            update2.append(value)
            
        score += update2[len(update2)//2]
        
print(score)
```
